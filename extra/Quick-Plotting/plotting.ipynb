### Quick Plotting
> Embedding Neuronal Constraints Drive Superior Learning in Recurrent Neural Networks

```python
import os #type: ignore
import glob #type: ignore
import pandas as pd #type: ignore
import matplotlib.pyplot as plt #type: ignore
from tqdm import tqdm #type: ignore
import numpy as np #type: ignore

plt.style.use('seaborn-v0_8-colorblind')

# Get the current working directory
# Note: This assumes the script is run from the 'simu' directory.
# If not, you'll need to provide the absolute path to the 'simu' directory.
cwd = os.getcwd()

config_files = glob.glob(os.path.join(cwd, '**', 'config.txt'), recursive=True)
metric_files = glob.glob(os.path.join(cwd, '**', 'metrics.csv'), recursive=True)
metric_sd_files = glob.glob(os.path.join(cwd, '**', 'metrics_SD.csv'), recursive=True)

config_data = {}
for config_file in config_files:
    # Extract the folder name from the file path
    folder_name = os.path.basename(os.path.dirname(config_file))
    with open(config_file, 'r') as f:
        config_data[folder_name] = f.read()
        
metrics_data = {}
for metric_file in metric_files:
    # Extract the folder name from the file path
    folder_name = os.path.basename(os.path.dirname(metric_file))
    metrics_data[folder_name] = pd.read_csv(metric_file)

metrics_sd_data = {}
for metric_sd_file in metric_sd_files:
    # Extract the folder name from the file path
    folder_name = os.path.basename(os.path.dirname(metric_sd_file))
    metrics_sd_data[folder_name] = pd.read_csv(metric_sd_file)

# print(config_data['6_6_2_sttc-precision'])
# print(metrics_data['6_6_2_sttc-precision'])
# print(metrics_sd_data['6_6_2_sttc-precision'])
```

```python
def plot_metrics(df, df_sd, task, folder):
    # Means
    acc_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'Accuracy')].iloc[0][2:]
    val_acc_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'Validation_Accuracy')].iloc[0][2:]
    modularity_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'Modularity')].iloc[0][2:]
    small_worldness_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'SmallWorldness')].iloc[0][2:]
    assortativity_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'Assortativity')].iloc[0][2:]
    entropy_row = df[(df['Task'] == task[0]) & (df['Metric'] == 'Entropy')].iloc[0][2:]

    # Standard deviations
    acc_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'Accuracy_SD')].iloc[0][2:]
    val_acc_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'Validation_Accuracy_SD')].iloc[0][2:]
    modularity_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'Modularity_SD')].iloc[0][2:]
    small_worldness_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'SmallWorldness_SD')].iloc[0][2:]
    assortativity_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'Assortativity_SD')].iloc[0][2:]
    entropy_sd_row = df_sd[(df_sd['Task'] == task[0]) & (df_sd['Metric'] == 'Entropy_SD')].iloc[0][2:]

    fig, axs = plt.subplots(3, 2, figsize=(14, 12), dpi=300)
    axs = axs.flatten()

    # Config info
    config_lines = config_data[folder].strip().split('\n')
    config_lines = [line for line in config_lines if line.strip()]
    config_lines = [line for line in config_lines if not line.startswith('save_folder_name')]
    axs[0].axis('off')
    axs[0].text(0, 1, '\n'.join(config_lines), fontsize=12, va='top', ha='left')

    # Accuracy (+ SD) - Grouped Bar Chart
    x = np.arange(len(acc_row.index))
    width = 0.35
    axs[1].bar(x - width/2, acc_row.values * 100, width, label='Train', yerr=acc_sd_row.values * 100, capsize=3)
    axs[1].bar(x + width/2, val_acc_row.values * 100, width, label='Validation', yerr=val_acc_sd_row.values * 100, capsize=3)
    axs[1].set_title('Accuracy (%)')
    axs[1].set_xlabel('Model')
    axs[1].set_ylabel('Accuracy (%)')
    axs[1].set_xticks(x)
    axs[1].set_xticklabels(acc_row.index, rotation=45, ha="right")
    axs[1].legend()
    axs[1].grid(axis='y', linestyle='--', alpha=0.7)

    # Modularity (+ SD) - Bar Chart
    axs[2].bar(modularity_row.index, modularity_row.values, yerr=modularity_sd_row.values, capsize=3)
    axs[2].set_title('Modularity')
    axs[2].set_xlabel('Model')
    axs[2].set_ylabel('Modularity')
    axs[2].tick_params(axis='x', rotation=45)
    axs[2].grid(axis='y', linestyle='--', alpha=0.7)

    # Small Worldness (+ SD) - Bar Chart
    axs[3].bar(small_worldness_row.index, small_worldness_row.values, yerr=small_worldness_sd_row.values, capsize=3)
    axs[3].set_title('Small Worldness')
    axs[3].set_xlabel('Model')
    axs[3].set_ylabel('Small Worldness')
    axs[3].tick_params(axis='x', rotation=45)
    axs[3].grid(axis='y', linestyle='--', alpha=0.7)

    # Assortativity (+ SD) - Bar Chart
    axs[4].bar(assortativity_row.index, assortativity_row.values, yerr=assortativity_sd_row.values, capsize=3)
    axs[4].set_title('Assortativity')
    axs[4].set_xlabel('Model')
    axs[4].set_ylabel('Assortativity')
    axs[4].tick_params(axis='x', rotation=45)
    axs[4].grid(axis='y', linestyle='--', alpha=0.7)

    # Entropy (+ SD) - Bar Chart
    axs[5].bar(entropy_row.index, entropy_row.values, yerr=entropy_sd_row.values, capsize=3)
    axs[5].set_title('Entropy')
    axs[5].set_xlabel('Model')
    axs[5].set_ylabel('Entropy')
    axs[5].tick_params(axis='x', rotation=45)
    axs[5].grid(axis='y', linestyle='--', alpha=0.7)

    fig.suptitle(task[1], fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.savefig(f'{folder}/{task[0]}.png', dpi=300)
    plt.close(fig)
```

```python
for folder in tqdm(metrics_data):
    df = metrics_data[folder]
    df_sd = metrics_sd_data[folder]
    plot_metrics(df, df_sd, ('Task1', 'One-choice inference'), folder)
    plot_metrics(df, df_sd, ('Task2', 'Go/NoGo'), folder)
    plot_metrics(df, df_sd, ('Task3', 'Perceptual decision-making'), folder)
```
